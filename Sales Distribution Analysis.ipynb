```python
# Importing and exploring data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Car_sales.csv")

# Display the first few rows
print(df.head())

```

```python
# Understanding the structure of our dataset

# Get basic information about the dataset
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Handling missing value
# 1. Remove rows where __year_resale_value  is missing (your main variable)
df = df.dropna(subset=['__year_resale_value'])
```

```python
print(df.isnull().sum())
```

```python
# 2. Fill numerical columns with median (handles price outliers)
df['Price_in_thousands'].fillna(df['Price_in_thousands'].median(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Handling outliers

df['Horsepower'].fillna(df['Horsepower'].median(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
#Fill missing numbers with the MIDDLE value (median)
print("\n" + "="*50)
print(" Fill missing numbers with median")
print("="*50)

# List of number columns to fix
number_columns = [
    'Engine_size',
    'Wheelbase', 
    'Width',
    'Length',
    'Curb_weight',
    'Fuel_capacity',
    'Fuel_efficiency',
    'Power_perf_factor',
    
]

# Fill each column with its middle value
for column in number_columns:
    if column in df.columns:  # Check if column exists
        middle_value = df[column].median()
        df[column] = df[column].fillna(middle_value)
        print(f"‚úì Fixed {column}")

# Step 4: Fill missing text with 'Unknown'
print("\n" + "="*50)
print(" Fill missing text with 'Unknown'")
print("="*50)

text_columns = ['Vehicle_type', 'Model', 'Manufacturer']

for column in text_columns:
    if column in df.columns:
        df[column] = df[column].fillna('Unknown')
        print(f"‚úì Fixed {column}")

# Step 5: Check if all fixed!
print("\n" + "="*50)
print(" Final check - Any missing values left?")
print("="*50)
print(df.isnull().sum())

if df.isnull().sum().sum() == 0:
    print("\n‚úÖ Success! No missing values!")
else:
    print("\n‚ö†Ô∏è Some values still missing")

# Step 6: Save the clean data
df.to_excel('car_sales_clean.xlsx', index=False)
print("\nüíæ Clean data saved as 'car_sales_clean.xlsx'")

print("\n" + "="*50)
print("ALL DONE! üéâ")
print("="*50)
```

```python
# Scatter plot of 'Sales in thousands' vs. 'Price in thousands'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price_in_thousands', y='Sales_in_thousands', data=df)
plt.title('Sales in thousands vs. Price in thousands')
plt.show()
```

```python
# Distribution of sale prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales_in_thousands'], kde=True)
plt.title('Distribution of Sales in thousands')
plt.show()
```

```python
# Grouping and Aggregation Statistics

# Average sale price by manufacturer
avg_price_by_manufacturer = df.groupby('Manufacturer')['Price_in_thousands'].mean().sort_values(ascending=False)
print("Average sale price by manufacturer:\n", avg_price_by_manufacturer)
```

```python
# Assuming Latest Launch represents the purchase date, extracting year from it
df['Year'] = df['Latest_Launch'].dt.year
sales_by_year = df.groupby('Year')['Sales_in_thousands'].sum()
plt.figure(figsize=(12, 6))
sales_by_year.plot(kind='bar')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales (in thousands)')
plt.show()
```

```python
# Top 10 best-selling models
top_models = df['Model'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_models.plot(kind='bar')
plt.title('Top 10 Best-Selling Models')
plt.xlabel('Model')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Exploring fundamental statistical measures that provide insights into our Car Sales dataset

# Measures of Central Tendency
mean_values = df[numeric_columns].mean()
median_values = df[numeric_columns].median()
mode_values = df[numeric_columns].mode().iloc[0]

print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nMode values:\n", mode_values)
```

```python
# Measures of Dispersion
variance_values = df[numeric_columns].var()
std_dev_values = df[numeric_columns].std()
range_values = df[numeric_columns].max() - df[numeric_columns].min()

print("\nVariance values:\n", variance_values)
print("\nStandard Deviation values:\n", std_dev_values)
print("\nRange values:\n", range_values)
```

```python
# Measures of Distribution Shape
skewness_values = df[numeric_columns].skew()
kurtosis_values = df[numeric_columns].kurtosis()

print("\nSkewness values:\n", skewness_values)
print("\nKurtosis values:\n", kurtosis_values)
```

```python
# Percentiles and Quartiles
percentiles = df[numeric_columns].quantile([0.25, 0.5, 0.75])
print("\nPercentiles:\n", percentiles)
```

```python
from scipy import stats

# Confidence Intervals for mean sale price
sale_price_mean = df['Price_in_thousands'].mean()
sale_price_std = df['Price_in_thousands'].std()
confidence_level = 0.95
degrees_freedom = df['Price_in_thousands'].count() - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sale_price_mean, sale_price_std/np.sqrt(df['Price_in_thousands'].count()))

print("\nConfidence interval for mean sale price:\n", confidence_interval)
```
